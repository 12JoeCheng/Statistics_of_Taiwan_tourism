資料來源：政府資料開放平臺

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

##

設定中文字體，matplotlib本身不支援，需要另外設定

```python
plt.rcParams['font.sans-serif']=['Noto Sans TC']
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_csv('https://stat.taiwan.net.tw/data/opendata/12')
#df = pd.read_csv('歷年來台旅客統計.csv')
df
```

##

將資料內的 '-' 字串設定成 NaN，方便資料的清理

```python
df.replace('-', np.NaN, inplace = True)
```

```python
df
```

##

將資料轉換成 int 與 float

```python
df['總計人數'] = df['總計人數'].str.replace(',', '')
df['外籍旅客人數'] = df['外籍旅客人數'].str.replace(',', '')
df['華僑旅客人數'] = df['華僑旅客人數'].str.replace(',', '')
```

```python
df['總計人數'] = df['總計人數'].astype(int)
df['外籍旅客人數'] = df['外籍旅客人數'].astype(int)
df['華僑旅客人數'] = df['華僑旅客人數'].astype(int)
df['總計成長率'] = df['總計成長率'].astype(float)
df['外籍旅客成長率'] = df['外籍旅客成長率'].astype(float)
df['華僑旅客成長率'] = df['華僑旅客成長率'].astype(float)
```

```python
df.info()
```

##

<font size = 5>外籍旅客與華僑旅客歷年人數</font>

```python
plt.style.use('ggplot')
fig, ax1 = plt.subplots(1,1, figsize = (14,10))


year = df['年別']
total_volume = df['總計人數']
foreign = df['外籍旅客人數']
overseas = df['華僑旅客人數']


ax1.plot(year, foreign, label = '外籍旅客', color = 'green')
ax1.plot(year, overseas, label = '華僑旅客', color = 'blue')
ax1.bar(year, total_volume, label = '總計', color = 'lightblue')

ax1.set_title('歷年來台旅客人數', fontsize = 20)
ax1.set_ylabel('人數（百萬）', fontsize = 13)
plt.legend(loc = 'best', fontsize = 13)
plt.tight_layout()
plt.show()
```

2020年末，新冠肺炎襲捲全球的緣故，造成隔年（2021）旅台人數急速下降。

##

<font size = 5>外籍旅客與華僑旅客歷年成長率</font>

```python
plt.style.use('ggplot')
fig, ax2 = plt.subplots(1,1, figsize = (14,10))

total_rate = df['總計成長率']
foreign_rate = df['外籍旅客成長率']
overseas_rate = df['華僑旅客成長率']

ax2.plot(year, foreign_rate, label = '外籍旅客', color = 'green')
ax2.plot(year, overseas_rate, label = '華僑旅客', color = 'blue')
ax2.bar(year, total_rate, label = '總體成長率', color = 'lightblue')

plt.axhline(0, color = 'black', linewidth = 1)

ax2.set_title('歷年來台旅客成長率', fontsize = 20)
ax2.set_ylabel('成長率（％）', fontsize = 13)
plt.legend(loc = 'best', fontsize = 13)
plt.tight_layout()
plt.show()
```

從圖表可以得出：相比於外籍旅客較穩定的成長率，華僑旅客的波動浮動較大，但兩者總體成長率大致上維持正成長。

##

<font size = 5>外籍旅客與華僑旅客歷年總計占比</font>

```python
plt.style.use('ggplot')
fig, (ax3, ax4) = plt.subplots(2,1, figsize = (14,10), sharex = True, sharey = True)

foreign_pct = df['外籍旅客占總計百分比']
overseas_pct = df['華僑旅客占總計百分比']

ax3.plot(year, foreign_pct, label = '外籍旅客', color = 'green')
ax4.plot(year, overseas_pct, label = '華僑旅客', color = 'blue')

ax3.axhline(80, color = 'black', label = '80%', linewidth = 1)
ax4.axhline(10, color = 'black', label = '10%', linewidth = 1)

ax3.set_title('外籍旅客與華僑旅客歷年總計占比', fontsize = 20)
ax3.set_ylabel('佔比（％）', fontsize = 13, loc = 'bottom')
ax3.legend(loc = 'best', fontsize = 13)
ax4.legend(loc = 'best', fontsize = 13)
plt.tight_layout()
plt.show()
```

外籍旅客歷年佔比皆大致維持在八成左右，只有在2010與2020間下降到大約五成; 華僑旅客則剛好相反，歷年維持在兩成左右，在2010與2020間上升到大約6成。

##

<font size = 3>從上述結果推論來看，外籍旅客常年維持較高的總人數, 正成長與高佔比下，對觀光產業佔比不低的台灣來說是一種穩定且相對強勁的收益來源。 另一方面，華僑則相對力道與影響力則相對弱，在各方面強過外籍旅客的時期大約為2015,16年左右，可能與華僑回國投票有關，2016年為台灣總統副總統及立法委員選舉，也是首屆有女性候選人參與總統競選。</font>
